IMPORTING LIBRARIES

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('Breast Cancer.csv')
```

VISUALLIZATION OF NULL VALUES IN HEATMAP

```python
sns.heatmap(data.isnull())
```

REMOVING ID AND UNNAMED COLUMN AS IT IS NOT REQUIRED

```python
data.drop(['Unnamed: 32','id'], axis=1,inplace=True)
data.head()
```

WE NEED THE DIAGNOSIS VALUE IN NUMERICAL FOR LIKE M=1 AND B=0

```python
data.diagnosis= [1 if value=='M' else 0 for value in data.diagnosis]
```

```python
data.head()
```

```python
data['diagnosis'].value_counts().plot(kind='bar')
```

NOW WE TAKE THE TARGET IN Y AND REST COLUMN IN X

```python
y= data['diagnosis']
x=data.drop(['diagnosis'],axis=1)
```

```python
y
```

```python
x
```

NOW WE NORMALIZE THE X DATA USING StandardScaler

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

x_scaled= scaler.fit_transform(x)
```

```python
x_scaled
```

SPLITING THE DATA 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test =train_test_split(x_scaled,y, test_size =0.3 , random_state= 42)
```

NOW TRAINING THE MODE BY lOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

# traint the model on the trainig data
lr.fit(x_train,y_train)

# predictig the target variable on test data
y_pred = lr.predict(x_test)
```

```python
# this is predicted value of y with repect to x_test
y_pred
```

```python
#And this is the actual value of y _test
y_test
```

TO CHECK THE ACCURACY OF THE MODEL

```python
from sklearn.metrics import accuracy_score

accuracy = accuracy_score(y_test,y_pred)
print(f" Accurany: {accuracy:.2f}")
```

CONFUSION MATRIX

```python
from sklearn.metrics import confusion_matrix
mat1 = confusion_matrix(y_pred,y_test)
```

```python
mat1
```

DISPLAY THE CONFUSION MATRIX

```python
plt.figure(figsize=(5,6))

sns.heatmap(mat1, annot=True, )

# Labels and title
plt.title('Confusion Matrix')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred))
```

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Range of values for z
z = np.linspace(-10, 10, 200)
sig = sigmoid(z)
```

```python
plt.figure(figsize=(8, 5))
plt.plot(z, sig, color='blue', label='Sigmoid Function')
plt.axhline(y=0.5, color='red', linestyle='--', label='Threshold = 0.5')
plt.axvline(x=0, color='gray', linestyle='--')
plt.title('Sigmoid Function Curve')
plt.xlabel('z')
plt.ylabel('Sigmoid(z)')
plt.legend()
plt.grid(True)
plt.show()
```
